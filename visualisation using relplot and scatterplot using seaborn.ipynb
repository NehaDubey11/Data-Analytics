```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('fivethirtyeight')
```

```python
data=sns.load_dataset('tips')
```

```python
data.head()
```

```python
sns.relplot(x='total_bill',y='tip',kind='scatter',data=data)
```

The hue parameter

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',data=data)
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',data=data)
```

```python
#using size parameter
```

```python
sns.relplot(x='total_bill',y='tip',hue='smoker',style='sex',size='size',sizes=(15,100),data=data)

#size can take only numerical coloumn not categorical.
```

Using sactterplot

```python
sns.scatterplot(x='total_bill',y='tip',hue='smoker',data=data)
```
